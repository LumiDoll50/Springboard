## Student Study Performance


### Story Overview
This story aims to uncover how different socio-demographic factors influence students' academic performance in math, reading, and writing. By examining variables such as gender, race/ethnicity, parental education, lunch status, and test preparation, the insights garnered can shape policies and interventions aimed at boosting student success.

### Audience: Education Policy Makers and School Administrators

### Questions to Ask:
1. **Gender Performance Gap**: How do male and female students compare in math, reading, and writing scores?
2. **Ethnic Disparities**: Are there any significant performance differences across various racial/ethnic groups?
3. **Impact of Parental Education**: How does the educational level of parents affect students' academic performance?
4. **Lunch Program**: Does having access to a standard vs. free/reduced lunch impact test scores?
5. **Test Preparation**: What is the effect of completing a test preparation course on the scores?


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df = pd.read_csv('study_performance.csv')
```

```python
df.head()
```

Educational equity and effective resource allocation in schools is important. Let's take a look at the features included in this study so that we can better understand the underlying factors and improve student outcomes.


### Gender-Based Performance

```python
# Calculate mean scores for male and female students in each subject
mean_scores = df.groupby('gender')[['math_score', 'reading_score', 'writing_score']].mean()

# Plot the mean scores
mean_scores.T.plot(kind='bar')
plt.title('Mean Scores by Gender in Each Subject')
plt.ylabel('Mean Score')
plt.xlabel('Subject')
plt.xticks(rotation=0)
plt.legend(title='Gender', bbox_to_anchor=(0.5, 1.15), loc='center', ncol=len(mean_scores.columns))
plt.show()
```

We see males outperform females in math but females outperform males in reading and writing.
Let's take a more detailed look:

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

sns.boxplot(ax=axes[0], data=df, x='gender', y='math_score')
axes[0].set_title('Math Scores by Gender')
axes[0].set_xlabel('Gender')
axes[0].set_ylabel('Math Score')

sns.boxplot(ax=axes[1], data=df, x='gender', y='reading_score')
axes[1].set_title('Reading Scores by Gender')
axes[1].set_xlabel('Gender')
axes[1].set_ylabel('Reading Score')

sns.boxplot(ax=axes[2], data=df, x='gender', y='writing_score')
axes[2].set_title('Writing Scores by Gender')
axes[2].set_xlabel('Gender')
axes[2].set_ylabel('Writing Score')

plt.tight_layout()
plt.show()
```

The data on math scores by gender shows that both female and male students have nearly identical median scores and interquartile ranges (IQRs), indicating similar central tendencies and variability in performance. Although the range of scores is comparable between genders, female students exhibit a slightly higher maximum score, as seen in the upper whisker. Additionally, there are more low-end outliers among female students, illustrating a few significantly lower scores within this group.

In contrast, reading and writing scores reveal that female students generally outperform their male counterparts. The median reading and writing scores for females are higher, suggesting better overall performance in these areas. The IQR for female reading scores is broader, indicating more variability, whereas the male IQR is slightly larger for writing scores. Both genders have similar whisker lengths, but females showcase a tendency for more low-end outliers, especially in reading and writing. Despite these outliers, the overall score distribution for females remains higher across both subjects.

```python
# Checking for significant disparities
gender_difference = df.groupby('gender')[['math_score', 'reading_score', 'writing_score']].mean()
gender_difference = gender_difference.loc['female'] - gender_difference.loc['male']

# Create a bar chart
plt.figure(figsize=(10, 6))
gender_difference.plot(kind='bar', color=['purple' if x > 0 else 'blue' for x in gender_difference])
plt.title('Score Differences Between Female and Male Students')
plt.xlabel('Subjects')
plt.ylabel('Score Difference (Female - Male)')

# Show values on bars
for i, v in enumerate(gender_difference):
    plt.text(i, v + (1 if v > 0 else -1), f'{v:.2f}', ha='center', va='center', color='black')

plt.axhline(0, color='black', linewidth=0.8)  # Add a horizontal line at y=0 for reference
plt.tight_layout()
plt.show()
```

The provided bar chart illustrates the score differences between female and male students across three subjects: math, reading, and writing. In math, male students outperform female students, with female students scoring on average 5 points lower than their male counterparts. This negative difference is depicted with a blue bar extending below the horizontal axis.

Conversely, female students outperform male students in both reading and writing. The chart shows reading scores with a purple bar above the horizontal axis, indicating that female students scored, on average, 7.14 points higher than male students. Similarly, for writing, the purple bar above the axis reveals that female students scored approximately 9.16 points higher on average than their male peers. Thus, while male students excel in math, female students show higher proficiency in reading and writing based on the average scores.

### Racial/Ethnic Performance Trends

```python
# Mean scores by race/ethnicity
race_ethnicity_means = df.groupby('race_ethnicity')[['math_score', 'reading_score', 'writing_score']].mean()

race_ethnicity_means.plot(kind='bar', figsize=(10,6))
plt.title('Average Scores by Race/Ethnicity')
plt.xlabel('Race/Ethnicity')
plt.ylabel('Average Scores')
plt.legend(title='Subject', bbox_to_anchor=(0.5, 1.15), loc='center', ncol=3)
plt.show()
```

Reading scores outperform math and writing scores consistently across all racial/ethnic groups, demonstrating a clear strength in this subject. Writing scores generally sit between the higher reading scores and the lower math scores for most groups, although in some groups, math scores are very close to writing scores.

Across all subjects, Groups D and E exhibit a marginally higher performance in comparison to Groups A, B, and C. Despite the variations in individual subject performance, this trend of higher achievement by Groups D and E is consistent across reading, math, and writing scores.

### Parental Education Influence

```python
# Mapping the parental level of education to a numerical scale for correlation analysis
education_levels = {'some high school': 1, 'high school': 2, 'some college': 3, "associate's degree": 4, "bachelor's degree": 5, "master's degree": 6}
df['parental_level_of_education_mapped'] = df['parental_level_of_education'].map(education_levels)

# Correlation analysis
correlation_matrix = df[['parental_level_of_education_mapped', 'math_score', 'reading_score', 'writing_score']].corr()

# Scatter plots with trend lines
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

sns.scatterplot(ax=axes[0], data=df, x='parental_level_of_education_mapped', y='math_score')
sns.regplot(ax=axes[0], data=df, x='parental_level_of_education_mapped', y='math_score', scatter=False, color='red')
axes[0].set_title('Parental Education vs Math Score')
axes[0].set_xlabel('Parental Education Level')
axes[0].set_ylabel('Math Score')

sns.scatterplot(ax=axes[1], data=df, x='parental_level_of_education_mapped', y='reading_score')
sns.regplot(ax=axes[1], data=df, x='parental_level_of_education_mapped', y='reading_score', scatter=False, color='red')
axes[1].set_title('Parental Education vs Reading Score')
axes[1].set_xlabel('Parental Education Level')
axes[1].set_ylabel('Reading Score')

sns.scatterplot(ax=axes[2], data=df, x='parental_level_of_education_mapped', y='writing_score')
sns.regplot(ax=axes[2], data=df, x='parental_level_of_education_mapped', y='writing_score', scatter=False, color='red')
axes[2].set_title('Parental Education vs Writing Score')
axes[2].set_xlabel('Parental Education Level')
axes[2].set_ylabel('Writing Score')

plt.tight_layout()
plt.show()
```

The relationship between parental education levels and student performance in math, reading, and writing shows a positive correlation across all subjects, as indicated by the positively sloping red regression lines in the scatter plots. Each blue dot, representing individual student scores, exhibits considerable scatter, suggesting that while higher parental education levels generally correspond to higher student scores, other factors also significantly contribute to student performance.

### Lunch Program Impact

Below we have a stacked bar chart visualizing the average scores in math, reading, and writing for students based on their lunch program, specifically comparing students who receive free/reduced lunches versus those who receive standard lunches.

```python
# Group by lunch type and calculate mean scores
lunch_stats = df.groupby('lunch')[['math_score', 'reading_score', 'writing_score']].mean()

# Stacked bar chart
lunch_stats.plot(kind='bar', stacked=True)
plt.title('Average Scores by Lunch Program')
plt.xlabel('Lunch Program')
plt.ylabel('Average Scores')
plt.legend(title='Subject', bbox_to_anchor=(0.5, 1.15), loc='center', ncol=3)
plt.show()
```

The data reveals that students enrolled in the free/reduced lunch program have lower average scores in math, reading, and writing compared to their peers who participate in the standard lunch program. The observations indicate that the overall average score for the free/reduced lunch group is significantly lower, as shown by the shorter total height of the bar in a stacked bar chart for math (blue), reading (orange), and writing (green) scores.

The overall performance clearly demonstrates that standard lunch program participants have superior academic achievements compared to their counterparts in the free/reduced lunch program.

### Effectiveness of Test Preparation Courses

Here we again use boxplots to illustrate the impact of test preparation courses on student scores in math, reading, and writing. Each plot compares scores for students who have completed a test preparation course with those who have not.

```python
test_prep_means = df.groupby('test_preparation_course')[['math_score', 'reading_score', 'writing_score']].mean()

# Side-by-side box plots for test preparation impact
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

sns.boxplot(ax=axes[0], data=df, x='test_preparation_course', y='math_score')
axes[0].set_title('Test Preparation Impact on Math Score')
axes[0].set_xlabel('Test Preparation')
axes[0].set_ylabel('Math Score')

sns.boxplot(ax=axes[1], data=df, x='test_preparation_course', y='reading_score')
axes[1].set_title('Test Preparation Impact on Reading Score')
axes[1].set_xlabel('Test Preparation')
axes[1].set_ylabel('Reading Score')

sns.boxplot(ax=axes[2], data=df, x='test_preparation_course', y='writing_score')
axes[2].set_title('Test Preparation Impact on Writing Score')
axes[2].set_xlabel('Test Preparation')
axes[2].set_ylabel('Writing Score')

plt.tight_layout()
plt.show()
```

Students who completed a test preparation course generally achieve higher scores across all subjectsmath, reading, and writingcompared to those who did not as exhibited by the higher median scores. While the whiskers, representing the range of scores, are similar for both groups, those who completed the course tend to score higher overall. Furthermore, the narrowing of the Interquartile Range (IQR) for the prepared group in math and writing suggests reduced variability in their scores, showcasing more consistent outcomes in their performance.

Additionally, there is a notable difference in the presence of low-end outliers between the two groups. Students who did not complete the test preparation course have more low-end outliers across all subjects, indicating a greater number of significantly lower scores compared to their peers who completed the course. This further enhances the insight that test preparation courses not only boost median scores but also contribute to more homogeneous and higher overall academic performance.

## SUMMARY

The study reveals crucial insights into the various socio-demographic factors affecting students' academic performances in math, reading, and writing. Findings underscore:
- Pronounced gender disparities, with female students outperforming male students in reading and writing, while male students slightly edge out female students in math.
- Significant ethnic disparities with reading emerging as a relative strength across all racial/ethnic groups and math being the weakest subject overall.
- Higher parental education levels correlate positively with better academic performance, although variability suggests additional influencing factors.
- Socioeconomic status, as indicated by the lunch program, significantly impacts academic outcomes, with students on standard lunch programs performing better than those on free/reduced programs.
-Test preparation courses are shown to effectively enhance student performance across all subjects.

These insights highlight the need for targeted interventions, such as academic support for economically disadvantaged students and broader access to test preparation courses. Education policymakers and school administrators are encouraged to consider these factors in their policy-making and resource allocation to foster an equitable and supportive learning environment for all students.
