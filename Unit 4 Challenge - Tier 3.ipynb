**2.2. Cleaning the data**

You might find you need to transpose your dataframe, check out what its row indexes are, and reset the index. You  also might find you need to assign the values of the first row to your column headings  . (Hint: recall the .columns feature of DataFrames, as well as the iloc[] method).

```python
# The Excel file's Metadata worksheet states that spatial coverage is by Regions and LA (Local Authority). We are only interested in the London Boroughs. Drop all region columns.
properties.drop(properties.iloc[:, 34:], axis = 1, inplace=True)

# Delete City of London: It is not a borough and not considered in this analysis.
del properties['City of London']

properties.info()
```

```python
# Transpose the DataFrame.
properties_T = properties.transpose()
properties_T.head()
```

```python
# Reset the index.
properties = properties_T.reset_index()   # Never have data (Borough) that is to be analyzed set as the index!
properties.head()
```

```python
# Reassign the columns the values in the row at index 0.
properties.columns = properties.iloc[0]
properties.head()
```

```python
# Remove the row at index 0.
properties = properties[1:]
properties.head()
```

**2.3. Cleaning the data (part 2)**

You might we have to **rename** a couple columns. How do you do this? The clue's pretty bold...

```python
# Rename columns.
properties = properties.rename(columns={'Unnamed: 0':'London Borough'})
properties.columns.values[1] = 'Local Authority Boundary'
properties.head()
```

**2.4.Transforming the data**

Remember what Wes McKinney said about tidy data? 

You might need to **melt** your DataFrame here. 

```python
# Melt, renaming Month and Average Price columns.
properties = properties.melt(id_vars=['London Borough', 'Local Authority Boundary'], var_name=['Month'], value_name='Average Price')
properties.head()
```

```python
# Check datatypes.
properties.dtypes
```

Remember to make sure your column data types are all correct. Average prices, for example, should be floating point numbers... 

```python
# Convert Average Price to a float.
properties['Average Price'] = properties['Average Price'].astype(float)
properties.dtypes
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
# .groupby throws an error if 'local Authority Boundary' exists in properties. Delete.
del properties['Local Authority Boundary']
```

```python
# Use groupby to calculate the mean of each borough for each year.
dfg = properties.groupby(by=['London Borough', 'Year']).mean()
dfg.head(10)
```

```python
# Reset the index.
dfg = dfg.reset_index()
dfg.head()
```

```python
# Function for calculating ratio.
def create_price_ratio(borough):
    price_1998 = float(borough['Average Price'][borough['Year']=='1998'])
    price_2018 = float(borough['Average Price'][borough['Year']=='2018'])
    price_ratio = [price_2018 / price_1998]
    return (price_ratio)
```

```python
# Loop for iterating over boroughs in dfg DataFrame.
final = {}

for b in dfg['London Borough'].unique():
    borough = dfg[dfg['London Borough'] == b]
    final[b] = create_price_ratio(borough)

print(final)
```

```python
# Convert dictionary to a DataFrame and reset index.
df_ratios = pd.DataFrame(final)
df_ratios.head()
```

```python
# Transpose DataFrame and reset the index.
df_ratios_T = df_ratios.T
df_ratios = df_ratios_T.reset_index()
df_ratios.head()
```

```python
# Rename 'index' and '0' columns.
df_ratios.rename(columns={'index':'Borough', 0:'Ratio'}, inplace=True)
df_ratios.head()
```

```python
# Sort for top 10 boroughs with greatest changes in price.
top10 = df_ratios.sort_values(by='Ratio', ascending=False)
print(top10)
```

```python
# Subset for most expensive borough.
dfg.loc[dfg['Average Price'].idxmax()]
```

```python
ax = top10[['Borough','Ratio']].plot(kind='bar')
ax.set_xticklabels(top10.Borough)
```
